```python
#Exercises: Day 10
#Iterate 0 to 10 using for loop, do the same using while loop.
def ztocent():
    for i in range(11):
        print(i)

    i = 0
    while i < 11:
        print(i)
        i += 1

#Iterate 10 to 0 using for loop, do the same using while loop.
def centtoz():
    for i in range(10, -1, -1):
        print(i)

    i = 10
    while i >= 0:
        print(i)
        i -= 1
```

```python
#hash traingle
def hash_triangle(n):
    for i in range(1, n + 1):
        print('#' * i)

hash_triangle(4)
```

```python
#Use for loop to iterate from 0 to 100 and print only even numbers
def print_even_numbers():
    for i in range(101):
        if i % 2 == 0:
            print(i)
print_even_numbers()
```

```python
#Use for loop to iterate from 0 to 100 and print only odd numbers 
def print_odd_numbers():
    for i in range(101):
        if i % 2 != 0:
            print(i)

print_odd_numbers()
```

```python
#Exercises: Level 2
#Use for loop to iterate from 0 to 100 and print the sum of all numbers
def sum_of_numbers():
    i = 0
    sum = 0
    while i <= 100:
        sum += i
        i += 1
    print("The sum of all numbers from 0 to 100 is:", sum)
sum_of_numbers()
```

```python
"""
Use for loop to iterate from 0 to 100 and print the sum of all evens and the sum of all odds. 
The sum of all evens is 2550. And the sum of all odds is 2500.
"""
def sum_of_evens_and_odds():
    sum_evens = 0
    sum_odds = 0
    for i in range(101):
        if i % 2 == 0:
            sum_evens += i
        else:
            sum_odds += i
    print("The sum of all evens is:", sum_evens)
    print("The sum of all odds is:", sum_odds)

sum_of_evens_and_odds()
```

```python
#Exercises: Level 3
'''
Go to the data folder and use the countries.py file. Loop through the countries and extract all the
countries containing the word land.
'''
file_path = "countries.py"
file_object = open(file_path, 'r')

countries = file_object.readlines()

def land_in_countries(countries):
    land_countries = []
    for country in countries:
        if 'land' in country:
            land_countries.append(country)
    return sorted(land_countries)

print(land_in_countries(countries))
```

```python
#This is a fruit list, [â€˜bananaâ€™, â€˜orangeâ€™, â€˜mangoâ€™, â€˜lemonâ€™] reverse the order using loop.
def reverse_fruit_list():
    fruits = ['banana', 'orange', 'mango', 'lemon']
    reversed_fruits = []
    for i in range(len(fruits)-1, -1, -1):
        reversed_fruits.append(fruits[i])
    return reversed_fruits

print(reverse_fruit_list())
```

```python
"""
Go to the data folder and use the countries_data.py file. 
What are the total number of languages in the data
Find the ten most spoken languages from the data
Find the 10 most populated countries in the world 
"""
def dictionary_operations():
        with open("countries-data.py", "r") as file_object:# Read operations here
            countries_data = file_object.read()
            countries_data = eval(countries_data)  # Convert string representation of dictionary to actual dictionary
            #return(countries_data)
        
        languages_set = set()
        for country in countries_data:
            for language in country.get('languages', []):
                languages_set.add(language)
        total_languages = len(languages_set)
        print("Total number of languages:", total_languages)

        # Find the ten most spoken languages
        language_count = {}
        for country in countries_data:
            for language in country.get('languages', []):
                if language in language_count:
                    language_count[language] += 1
                else:
                    language_count[language] = 1

        # Get the ten most spoken languages
        most_spoken_languages = sorted(language_count.items(), key=lambda x: x[1], reverse=True)[:10]
        print("Ten most spoken languages:")
        for language, count in most_spoken_languages:
            print(f"{language}: {count}")

        # Find the 10 most populated countries
        most_populated_countries = sorted(countries_data, key=lambda x: x.get('population', 0), reverse=True)[:10]
        print("Ten most populated countries:")
        for country in most_populated_countries:
            print(f"{country['name']}: {country['population']}")

dictionary_operations()
```
