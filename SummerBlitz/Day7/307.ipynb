```python
#Exercises: Day 7 # sets

it_companies = {'Facebook', 'Google', 'Microsoft', 'Apple', 'IBM', 'Oracle', 'Amazon'}
A = {19, 22, 24, 20, 25, 26}
B = {19, 22, 20, 25, 26, 24, 28, 27, 30}
age = [22, 19, 24, 25, 26, 24, 25, 24]

```

```python
#Exercises: Level 1
# Find the length of the set it_companies
length = len(it_companies)
print("Length of it_companies set:", length)

# Add â€˜Twitterâ€™ to it_companies
it_companies.add('Twitter')
print("Updated it_companies set:", it_companies)

#Insert multiple IT companies at once to the it_companies
it_companies.update(['LinkedIn', 'Snapchat', 'Reddit'])
print("it_companies after adding multiple companies:", it_companies)

#Remove one of the companies from the set it_companies
it_companies.pop()
print("it_companies after removing one company:", it_companies)

#What is the difference between remove and discard
# remove() will raise a KeyError if the item to be removed is not found in the set.
# discard() will not raise an error if the item is not found; it simply does nothing.

```

```python
#Exercises: Level 2
#Join A and B
joined_set = A.union(B)
print("Joined set A and B:", joined_set)

#Find A intersection B
intersection_set = A.intersection(B)
print("Intersection of A and B:", intersection_set)

#Is A subset of B
is_subset = A.issubset(B)
print("Is A a subset of B?:", is_subset)

#Are A and B disjoint sets
are_disjoint = A.isdisjoint(B)
print("Are A and B disjoint sets?:", are_disjoint)
```

```python
#Join A with B and B with A
A_union_B = A.union(B)
B_union_A = B.union(A)
print("A union B:", A_union_B)
print("B union A:", B_union_A)

#Delete the sets completely
del A
del B
#print(A)  # This will raise an error since A is deleted
```

```python
#Exercises: Level 3
#Convert the ages to a set and compare the length of the list and the set, which one is bigger?
age_set = set(age)
print("Length of age list:", len(age))
print("Length of age set:", len(age_set))
if len(age) > len(age_set):
    print("The list is bigger.")
elif len(age) < len(age_set):
    print("The set is bigger.")
else:
    print("Both are of equal length.")

#Explain the difference between the following data types: string, list, tuple and set

#String are combinations of charchters used to represent text. They are ordered and immutable.
#List are ordered collections of items that are mutable, meaning they can be changed after creation.
#Tuple are similar to lists but are immutable, meaning once created, their elements cannot be changed.
#Set are unordered collections of unique items, meaning they do not allow duplicate values and are mutable

#I am a teacher and I love to inspire and teach people. How many unique words have been used in the sentence? Use the split methods and set to get the unique words.
sentence = "I am a teacher and I love to inspire and teach people."
words = sentence.split()
unique_words = set(words)
print("Number of unique words in the sentence:", len(unique_words)) 
```
