Lesson 14
Iterations: More Patterns

```python
def ask_number():

    total = 0
    ask = ""
    count = 0

    while ask != "Done":
        ask = input("Enter a number or 'Done' to finish: ")

        if ask == "Done":
            print(f"The total is: {total} and your average is: {total/count}")
            break
        else:
            print(f"You entered: {ask}")
            total += float(ask)
            count += 1

    print(f"The total is: {total}")

ask_number()
```
